### 코딩 시작

```python
# 음성 파일 및 음성 파일 이름 불러오기
folder = 'C:/Users/HONGHYEIN/Desktop/홍혜인/4학년 2학기/캡스톤/dev/학습 데이터/wav'
files = paths.list_files(folder)

dataset = []
name = []

for i in files:
    y, sr = librosa.load(i, sr = 16000)
    dataset.append([y, sr])
    name.append((os.path.basename(i)))
```

```python
# 음성 파일 길이 저장하기
duration = []

for data in dataset:
    duration.append(int(librosa.get_duration(data[0], data[1])))

# 음성 파일 1초씩 자르고 저장하기
folder = 'C:/Users/HONGHYEIN/Desktop/홍혜인/4학년 2학기/캡스톤/dev/학습 데이터/wav/1초/'

for data, name, duration in zip(dataset, name, duration):
    y, sr = data
    for i in range(duration):
        cutted_y = y[i*16000:(i+1)*16000]
        cutted_name = name[:-4]
        sf.write(folder + cutted_name + '_' + str(i+1) + '.wav', cutted_y, sr, 'PCM_16')
```
