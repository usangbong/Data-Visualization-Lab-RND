```python
# VAR causality test
def nearest_edge_gc_all(df, adj_info ,pk, index_num, time):
    # ==== adj_edge ==== #
    adj_edge = list(adj_info[adj_info['from'] == int(pk[0][index_num])]["to"])
    state_edge = pk[0][index_num]
    print('edge :',state_edge)
    if len(adj_edge) == 0:
        return print('Not exist adj_edge')
    
    else:
        for _ in range(len(adj_edge)):
            adj_edge[_] = str(adj_edge[_])
        adj_edge_df = df[adj_edge]
        
        if state_edge in list(adj_edge_df.columns):
            print(adj_edge_df.columns)
            print(adj_edge_df.columns.difference([state_edge]))
            model = VAR(adj_edge_df)
            result = model.fit(time)
            gc_multi = result.test_causality(state_edge,causing = list(adj_edge_df.columns.difference([state_edge])))
        
        else:
            adj_edge_list = list(adj_edge_df.columns)
            adj_edge_df[state_edge] = df[state_edge]
            model = VAR(adj_edge_df)
            result = model.fit(time)
            gc_multi = result.test_causality(state_edge,causing = adj_edge_list)
        print(gc_multi)
```

```python
# ì¸ì ‘ edgeë“¤ê³¼ ëª¨ë‘ gcê´€ê³„ì— ìžˆëŠ” edgeë“¤

def all_adj_edge_gc(df,adj_info,pk,time):
    adj_edge_gc_all = []
    for index_num in range(len(pk)):
        # ==== adj_edge ==== #
        state_edge = pk[index_num]
        adj_edge = list(adj_info[adj_info['from'] == int(pk[index_num])]["to"])
        
        if state_edge in adj_edge:
            adj_edge.remove(state_edge)
        # =================================#

        if len(adj_edge) == 0:
            return print('Not exist adj_edge')

        else:
            p_value_over = []
            for i in range(len(adj_edge)):
                # ======= granger causality test ========= #
                p_value = grangercausalitytests(df[[state_edge, str(adj_edge[i])]],[time],verbose = False)[12][0]['ssr_ftest'][1]
                if p_value > 0.05:
                    p_value_over.append(p_value)

        if len(p_value_over) == 0:
            adj_edge_gc_all.append(state_edge)
    return adj_edge_gc_all
```

```python
# ëª¨ë“  ì¸ì ‘ edgeë“¤ê³¼ gcê°€ ìžˆëŠ” edgeì˜ ì—­ë°©í–¥ gcì¤‘ ì´ìƒì¹˜ë“¤
def trans_gc(df, adj_info, edges, pk, time):
    dic = {}
    for state_edge in edges:
        not_gc = []
        adj_edge = list(adj_info[adj_info['from'] == int(state_edge)]["to"])
        if state_edge in adj_edge:
            adj_edge.remove(state_edge)
        p_value_over = []
        for i in range(len(adj_edge)):
            # ======= granger causality test ========= #
            p_value = grangercausalitytests(df[[str(adj_edge[i]), state_edge]],[time],verbose = False)[12][0]['ssr_ftest'][1]
            if p_value > 0.05:
                p_value_over.append(str(adj_edge[i]))
        dic[state_edge] = p_value_over
    return dic
```

```python
def not_gcs_count(df,pk,time):
    p_value_over = []
    
    for edge in pk[0]:
        edge_list = pk[0]
        
        for i in range(len(edge_list)):    
            p_value = grangercausalitytests(df[[edge,edge_list[i]]],[time],verbose = False)[12][0]['ssr_ftest'][1]
            if p_value >0.05:
                p_value_over.append(p_value)
            p_value = grangercausalitytests(df[[edge_list[i],edge]],[time],verbose = False)[12][0]['ssr_ftest'][1]
            if p_value >0.05:
                p_value_over.append(p_value)
    return len(p_value_over)
```
